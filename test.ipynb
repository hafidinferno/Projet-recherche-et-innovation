```python
from tensorflow import keras
import numpy as np
from PIL import Image, ImageOps
import tensorflow as tf
import cv2

model = tf.keras.models.load_model("./model.h5")
model2 = tf.keras.models.load_model("./model2.h5")
model3 = tf.keras.models.load_model("./model3.h5")
model4 = tf.keras.models.load_model("./m.h5")
model5 = tf.keras.models.load_model("./model5.h5")



```

```python
img = Image.open("./ID_001148.png").convert("L")
```

```python
img_test = ImageOps.equalize(img)
img = img_test.convert("RGB").resize([256,256])
img2 = img_test.convert("RGB").resize([128,128])
img = np.array(img)
img = np.expand_dims(img, axis=0)
img = img / 255.0
img2 = np.array(img2)
img2 = np.expand_dims(img2, axis=0)
img2 = img2 / 255.0
```

```python
result = model.predict(img)
result2 = model2.predict(img)
result3 = model3.predict(img)
result4 = model4.predict(img2)
result5 = model5.predict(img)
#result6 = model6.predict(img)
```
